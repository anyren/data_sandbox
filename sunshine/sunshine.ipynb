```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from unicodedata import normalize
```

```python
table_sunshine = pd.read_html('https://en.wikipedia.org/wiki/List_of_cities_by_sunshine_duration')
```

```python
len(table_sunshine)
```

```python
continents = ["Africa","Asia","Europe","North America","South America","Oceania"]
```

```python
all_sunshine = table_sunshine[2]
all_sunshine['Continent'] = continents[0]
c=1
for t in range(3,8):
    table_c = table_sunshine[t]
    table_c['Continent'] = continents[c]
    all_sunshine = pd.concat([all_sunshine,table_c])
    c+=1
```

```python
all_sunshine=all_sunshine.drop(columns='Ref.')
```

```python
col = all_sunshine.pop("Continent")
all_sunshine.insert(0, col.name, col)
```

```python
all_sunshine
```

```python
all_sunshine['Day'] = all_sunshine['Year']/365
```

```python
all_sunshine.to_csv('world_sunshine.csv', index=False)
```
